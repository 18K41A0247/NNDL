Find the global minimum point and value for the function f(x)=x^4+3*x^2+10 

* Manual calculation for two iterations

* Optimal solution using python programming



```
from sympy import *
x=symbols('x')
f=x**4+3*x**2+10
y=diff(f,x) 
print(y)
X=int(input("enter x value:"))
itr=1 
n=0.1 
epoch=2
while (itr <= epoch) :
   Y=y.subs(x,X)
   print(f"differentiation of f({X}) =",Y)
   deltax=-n*Y
   print("delta of x =",deltax)
   newx = X + deltax
   print("new x =",newx)
   itr= itr + 1
   X=newx
   print(' ')
print("final x value =",X)
print("Minimum value of f(x) is",f.subs(x,X))
```
